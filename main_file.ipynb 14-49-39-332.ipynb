### **Levene‚Äôs Test**

For each race segment, you are testing whether the variability (variance) in split times is significantly different between elite and non-elite runners. Here's how you can frame your null and alternative hypotheses:

- H‚ÇÄ (Null Hypothesis): There is no significant difference in the variability (variance) of pacing between elite and non-elite runners. Both groups have similar consistency in pacing.

- H‚ÇÅ (Alternative Hypothesis): Elite runners exhibit significantly lower variability (i.e., more consistency) in their pacing compared to non-elite runners.

#### **Running the test**


```python
# P-value threshold 
# Œ± < 0.05

split_columns = [
    'avg_split_5K_to_10K', 
    'avg_split_10K_to_15K', 
    'avg_split_15K_to_20K', 
    'avg_split_20K_to_25K',
    'avg_split_25K_to_30K', 
    'avg_split_30K_to_35K', 
    'avg_split_35K_to_40K'
]

# Loop through each segment
for column in split_columns:
    elite_data = df_h2_allyears[df_h2_allyears['Status'] == 'elite'][column]
    non_elite_data = df_h2_allyears[df_h2_allyears['Status'] == 'non-elite'][column]

    # Levene's Test for Equal Variance
    levene_test = stats.levene(elite_data, non_elite_data)
    
    # Calculate Standard Deviation (optional)
    elite_sd = elite_data.std()
    non_elite_sd = non_elite_data.std()

    print(f"Segment: {column}")
    print(f"Elite SD: {elite_sd}, Non-Elite SD: {non_elite_sd}")
    print(f"Levene Test p-value: {levene_test.pvalue}\n")
```

#### **Interpreting the results & Conclusion**   

Based on the Levene‚Äôs Test results, **none of the segments show a significant difference in variability** (p-value > 0.05 for all segments).

### T-Test

#### Set the hypothesis

**(H‚ÇÄ) Null Hypothesis** : There is no significant difference in the finish times between runners with negative splits and those without negative splits.
(i.e., Negative splits do not correlate with faster finish times.)

**(H‚ÇÅ)Alternative Hypothesis** : Runners with negative splits have significantly faster finish times than those without negative splits.
(i.e., Negative splits correlate with faster finish times.)

#### Prepare test

```python
#Separate data based on split type
negative_splits = df_h3[df_h3['split_type'] == 'Negative Split']['finish_time']
positive_splits = df_h3[df_h3['split_type'] == 'Positive Split']['finish_time']
```

#### Test

```python
# P-value threshold 
# Œ± < 0.05

# Convert to timedelta then to total seconds
df_h3['finish_time'] = pd.to_timedelta(df_h3['finish_time']).dt.total_seconds()

# Separate data based on split type
negative_splits = df_h3[df_h3['split_type'] == 'Negative Split']['finish_time']
positive_splits = df_h3[df_h3['split_type'] == 'Positive Split']['finish_time']

# Perform independent t-test
t_stat, p_value = ttest_ind(negative_splits, positive_splits, equal_var=False)

# Output the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

#### Results & Conclusion

- **t-test** results show a t-statistic of -30.67 and a p-value of 1.54e-185, which is an extremely small p-value, well below any conventional significance level (like 0.05 or 0.01). 

**This suggests:**

- **Significance**: The p-value being so close to zero indicates a highly significant difference between the two groups (runners with negative splits and positive splits).

- **Conclusion**: Given the p-value, we reject the null hypothesis, meaning there is a statistically significant difference in finish times between runners who achieved negative splits and those who did not.

This implies that runners with negative splits likely have a different (likely faster) finish time compared to those with positive splits, supporting the idea that negative splits might contribute to better overall performance.

```python
# Convert 'time' and 'pace' columns to timedelta
merged_h4['time_initial'] = pd.to_timedelta(merged_h4['time_initial'])
merged_h4['time_subsequent'] = pd.to_timedelta(merged_h4['time_subsequent'])
merged_h4['pace_mph_initial'] = pd.to_timedelta(merged_h4['pace_mph_initial'])
merged_h4['pace_mph_subsequent'] = pd.to_timedelta(merged_h4['pace_mph_subsequent'])
merged_h4['pace_kmh_initial'] = pd.to_timedelta(merged_h4['pace_kmh_initial'])
merged_h4['pace_kmh_subsequent'] = pd.to_timedelta(merged_h4['pace_kmh_subsequent'])

# Calculate time_diff and performance differences
merged_h4['time_diff'] = merged_h4['time_subsequent'] - merged_h4['time_initial']
merged_h4['pace_mph_diff'] = merged_h4['pace_mph_subsequent'] - merged_h4['pace_mph_initial']
merged_h4['pace_kmh_diff'] = merged_h4['pace_kmh_subsequent'] - merged_h4['pace_kmh_initial']


# Apply formatting to the timedelta columns
# Call function to format timedelta to 'HH:MM:SS' and handle negative values:'format_timedelta(td)'

merged_h4['time_initial'] = merged_h4['time_initial'].apply(format_timedelta)
merged_h4['time_subsequent'] = merged_h4['time_subsequent'].apply(format_timedelta)
merged_h4['time_diff'] = merged_h4['time_diff'].apply(format_timedelta)
merged_h4['pace_mph_diff'] = merged_h4['pace_mph_diff'].apply(format_timedelta)
merged_h4['pace_kmh_diff'] = merged_h4['pace_kmh_diff'].apply(format_timedelta)

# Remove '0 days' from pace columns (if any)
merged_h4['pace_mph_initial'] = merged_h4['pace_mph_initial'].astype(str).str.replace('0 days ', '')
merged_h4['pace_mph_subsequent'] = merged_h4['pace_mph_subsequent'].astype(str).str.replace('0 days ', '')
merged_h4['pace_kmh_initial'] = merged_h4['pace_kmh_initial'].astype(str).str.replace('0 days ', '')
merged_h4['pace_kmh_subsequent'] = merged_h4['pace_kmh_subsequent'].astype(str).str.replace('0 days ', '')

# Display the relevant columns
print(merged_h4[['name', 'time_initial', 'time_subsequent', 'time_diff', 'pace_mph_diff', 'pace_kmh_diff']].head())

```

### 5. Vizualization on Performance Changes

```python
# Ensure time columns are converted to Timedelta if they aren't already
time_columns = ['time_initial', 'time_subsequent']

for col in time_columns:
    if merged_h4[col].dtype == 'object':  # Check if the column is still a string
        merged_h4[col] = pd.to_timedelta(merged_h4[col])

# Calculate time difference in seconds
merged_h4['time_diff_sec'] = (merged_h4['time_subsequent'] - merged_h4['time_initial']).dt.total_seconds()

# Create a histogram with KDE overlay for the time difference in seconds
plt.figure(figsize=(10, 6))
sns.histplot(merged_h4['time_diff_sec'], kde=True, color='skyblue')
plt.title("Distribution of Time Difference (seconds)")
plt.xlabel("Time Difference (seconds)")
plt.ylabel("Frequency")
plt.show()
```

### 6. Checking Normality of Distribution: Shapiro-Wilk Test

```python
# Conduct normality test (Shapiro-Wilk)
print("Normality Test Results for Time Difference (seconds):")

# Shapiro-Wilk test
shapiro_stat, shapiro_p = shapiro(merged_h4['time_diff_sec'])
print(f"Shapiro-Wilk Test: W-statistic={shapiro_stat:.3f}, p-value={shapiro_p:.3f}")
if shapiro_p > 0.05:
    print("   The time difference is likely normal (fail to reject H0).")
else:
    print("   The time difference is likely not normal (reject H0).")
```

### 7. Conduct an approriate test for non-normal distribution: Wilcoxon Signed-Rank Test

```python
# Null hypothesis (H0): There is no significant difference between the initial and subsequent race times/pace (the difference is zero).
# Alternative hypothesis (H1): There is a significant difference between the initial and subsequent race times/pace (i.e., improvement).b
```

```python
# Conduct the Wilcoxon Signed-Rank Test for time difference in seconds
stat_time, p_time = wilcoxon(merged_h4['time_diff_sec'])
print("Wilcoxon Signed-Rank Test for Finishing Time Difference:")
print(f"Statistic={stat_time:.3f}, p-value={p_time:.3f}")
if p_time < 0.05:
    print("Significant improvement in finishing times (reject H0).")
else:
    print("No significant improvement in finishing times (fail to reject H0).")
```
