```python
# Setup cache and retry mechanism
cache_session = requests_cache.CachedSession('.cache', expire_after=3600)
retries = Retry(total=5, backoff_factor=0.2)
cache_session.mount("https://", HTTPAdapter(max_retries=retries))

# Define the API endpoint and parameters
url = "https://archive-api.open-meteo.com/v1/era5"
params = {
    "latitude": 42.355007,
    "longitude": -71.12906,
    "start_date": "2015-04-20",
    "end_date": "2015-04-20",
    "hourly": [
        "temperature_2m", "relative_humidity_2m", "dew_point_2m", 
        "apparent_temperature", "precipitation_probability", "precipitation", 
        "rain", "showers", "pressure_msl", "surface_pressure", "cloud_cover", 
        "cloud_cover_low", "cloud_cover_mid", "cloud_cover_high", "visibility", 
        "evapotranspiration", "wind_speed_10m", "wind_direction_10m", "wind_gusts_10m", 
        "soil_temperature_0cm", "soil_moisture_0_to_1cm", "is_day"
    ],
}

# Make the API request
response = cache_session.get(url, params=params)
data = response.json()

# Extract and format hourly data into a DataFrame
hourly_data = data.get("hourly", {})
api_15 = pd.DataFrame(hourly_data)  # Store the data in a new DataFrame called 'api_15'
api_15["time"] = pd.to_datetime(api_15["time"])

# Filter to only include the range from 10:00 to 18:00
api_15 = api_15.set_index("time").between_time("10:00", "18:00").reset_index()


api_15
```

# **Data Cleaning**

### df_all

```python
# Assign year to each dataframe and concatenate
df_all = pd.concat([df.assign(year=year) for df, year in zip([df_15, df_16, df_17], [2015, 2016, 2017])], ignore_index=True)

# Drop specified columns and create 'id' column
df_all = (df_all
               .drop(columns=['Citizen', 'Unnamed: 9', 'Unnamed: 8', 'State', 'Unnamed: 0'])
               .reset_index(drop=True))

# Create a new 'id' column with a unique identifier for each row
df_all['id'] = range(1, len(df_all) + 1)

# Reorder columns
df_all = df_all[['id', 'year', 'Bib'] + [col for col in df_all.columns if col not in ['id', 'year', 'Bib']]]


# Additional renaming to check for the correct column names
df_all.rename(columns={
    'Proj Time': 'proj_time',
    'Official Time': 'time'
}, inplace=True)


# Convert all other column names to lowercase except for distance columns
df_all.columns = [
    col if col in ['5K', '10K', '15K', '20K', '25K', '30K', '35K', '40K'] else col.lower()
    for col in df_all.columns
]

# Specify the columns to check for '-'
columns_to_check = ['5K', '10K', '15K', '20K', 'half', '25K', '30K', '35K', '40K']

# Drop rows where any of the specified columns contain '-'
df_all = df_all[~df_all[columns_to_check].isin(['-']).any(axis=1)]
```

```python
# Rename additional columns
df_all.rename(columns={
    'overall': 'overall_rank',
    'gender': 'gender_rank',
    'division': 'division_rank',
    'm/f': 'gender'  # Rename 'm/f' to 'gender'
}, inplace=True)

df_all.head()
```

```python
# Get unique duplicate names
unique_duplicates = df_all['name'][df_all.duplicated('name')].unique()

# Display unique duplicate names
print(unique_duplicates)
```
