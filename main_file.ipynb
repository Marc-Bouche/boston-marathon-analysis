# API

```python
# Your Meteomatics API credentials
username = 'ironhack_bouch_marc'  # My username
password = '9aIOqFi122'           # My password

# Define the endpoint
base_url = 'https://api.meteomatics.com/'

# Define the time zone for Boston
boston_tz = pytz.timezone('America/New_York')

# Define the specific start and end dates for each year
dates = [
    ("2015-04-20 10:00", "2015-04-20 16:00"),
    ("2016-04-18 10:00", "2016-04-18 16:00"),
    ("2017-04-17 10:00", "2017-04-17 16:00"),
]

# Loop through each date pair to make API requests
for start, end in dates:
    # Convert strings to datetime objects in local Boston time
    start_date = boston_tz.localize(datetime.strptime(start, '%Y-%m-%d %H:%M'))
    end_date = boston_tz.localize(datetime.strptime(end, '%Y-%m-%d %H:%M'))

    # Format the dates for the API request
    start_date_str = start_date.strftime('%Y-%m-%dT%H:%M:%S')
    end_date_str = end_date.strftime('%Y-%m-%dT%H:%M:%S')

    # Define the location (e.g., Boston)
    location = '42.3601,-71.0589'  # Latitude, Longitude for Boston

    # Query for specific weather data (e.g., temperature and precipitation)
    parameters = f'{start_date_str}/{end_date_str}/PT1H:temperature_2m,precipitation_sum'

    # Full URL
    url = f'{base_url}{parameters}/{location}/json'

    # Make the API request
    response = requests.get(url, auth=(username, password))

    # Check the response
    if response.status_code == 200:
        data = response.json()
        print(f"Weather data for {start_date_str} to {end_date_str}:")
        print(data)  # Print the weather data
    else:
        print(f'Error for {start_date_str} to {end_date_str}: {response.status_code} - {response.text}')
```
